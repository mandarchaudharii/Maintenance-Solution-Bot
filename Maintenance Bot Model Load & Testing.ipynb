```python
load_in_4bit = True  # Set it to True if you want to load the model with 4-bit quantization, or False otherwise

if True:
    # I highly do NOT suggest - use Unsloth if possible
    from peft import AutoPeftModelForCausalLM
    from transformers import AutoTokenizer

    load_in_4bit = True  # Define load_in_4bit here

    model = AutoPeftModelForCausalLM.from_pretrained(
        "mandarchaudharii/maintenancebot", # YOUR MODEL YOU USED FOR TRAINING
        load_in_4bit=load_in_4bit,
    )
    tokenizer = AutoTokenizer.from_pretrained("mandarchaudharii/maintenancebot")
```

```python
import warnings
warnings.filterwarnings('ignore')

alpaca_prompt = """
### Question:
{}

### Machine:
{}

### Response:
{}"""

print()
print("Hey There, Any Maintenance Issues Today ?")
print()

input_text = input("Tell me the Machine Name: ")

print()

instruction = input("Tell me the Issue: ")

print()

output_text = ""

formatted_input = alpaca_prompt.format(instruction, input_text, output_text)

inputs = tokenizer(formatted_input, return_tensors="pt").to("cuda")
outputs = model.generate(**inputs, max_new_tokens=128)

# Decode and print the generated text
print("Machine:", tokenizer.decode(outputs[0], skip_special_tokens=True))
```
